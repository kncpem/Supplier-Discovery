# Financial Score

```python
# Define the financial metrics
metrics = {
    "revenue_growth_rate": revenue_growth_rate,
    "gross_profit_margin": gross_profit_margin,
    "operating_profit_margin": operating_profit_margin,
    "net_profit_margin": net_profit_matgin,
    "operating_cash_flow": operating_cash_flow / total_revenue,
    "free_cash_flow": free_cash_flow / total_revenue,
    "debt_to_equity_ratio": debt_to_equity_ratio,
    "roe": return_to_equity_roe,
    "roa": return_on_assets_roa,
    "eps": earnings_per_share_eps,
    "pe_ratio": price_to_earnings_pe_ratio
}

# Define the normalization functions
def normalize_growth_rate(rate):
    if rate > 0.15:
        return 1
    elif rate > 0.10:
        return 0.8
    elif rate > 0.05:
        return 0.6
    else:
        return 0.4

def normalize_margin(margin):
    if margin > 0.40:
        return 1
    elif margin > 0.30:
        return 0.8
    elif margin > 0.20:
        return 0.6
    else:
        return 0.4

def normalize_cash_flow_to_revenue(ratio):
    if ratio > 0.15:
        return 1
    elif ratio > 0.10:
        return 0.8
    elif ratio > 0.05:
        return 0.6
    else:
        return 0.4

def normalize_debt_to_equity(ratio):
    if ratio < 0.5:
        return 1
    elif ratio < 1:
        return 0.8
    elif ratio < 2:
        return 0.6
    else:
        return 0.4

def normalize_return_on_metric(metric):
    if metric > 0.20:
        return 1
    elif metric > 0.15:
        return 0.8
    elif metric > 0.10:
        return 0.6
    else:
        return 0.4

def normalize_pe_ratio(ratio):
    if ratio < 10:
        return 1
    elif ratio < 15:
        return 0.8
    elif ratio < 20:
        return 0.6
    else:
        return 0.4

# Normalization mapping
normalization_functions = {
    "revenue_growth_rate": normalize_growth_rate,
    "gross_profit_margin": normalize_margin,
    "operating_profit_margin": normalize_margin,
    "net_profit_margin": normalize_margin,
    "operating_cash_flow": normalize_cash_flow_to_revenue,
    "free_cash_flow": normalize_cash_flow_to_revenue,
    "debt_to_equity_ratio": normalize_debt_to_equity,
    "roe": normalize_return_on_metric,
    "roa": normalize_return_on_metric,
    "eps": lambda x: 0.6,  # Assuming industry average normalization
    "pe_ratio": normalize_pe_ratio
}

# Weightings for each metric
weightings = {
    "revenue_growth_rate": 0.1,
    "gross_profit_margin": 0.1,
    "operating_profit_margin": 0.1,
    "net_profit_margin": 0.1,
    "operating_cash_flow": 0.1,
    "free_cash_flow": 0.1,
    "debt_to_equity_ratio": 0.1,
    "roe": 0.1,
    "roa": 0.1,
    "eps": 0.1,
    "pe_ratio": 0.1
}

# Calculate the normalized score
normalized_scores = {metric: normalization_functions[metric](value) for metric, value in metrics.items()}

# Calculate the weighted score
total_score = sum(normalized_scores[metric] * weight for metric, weight in weightings.items())*100

# Print the total score and its interpretation
print(f"Total Financial Score: {total_score:.2f}")

if total_score >= 90:
    print("Interpretation: Excellent")
elif total_score >= 80:
    print("Interpretation: Very Good")
elif total_score >= 70:
    print("Interpretation: Good")
elif total_score >= 60:
    print("Interpretation: Average")
elif total_score >= 50:
    print("Interpretation: Below Average")
else:
    print("Interpretation: Poor")
```
